```python
#identifies account with budget > 100,000 to print 
import json

with open('c:/users/wrigh/cgcapital.geojson') as f:
  data = json.load(f)
print('List of projects over 100,000 dollars')
print('_____________________________________' )
for property in data['features']:
   dict = property['properties'] 
#   print(dict)
   if dict['budgeted_amount'] > 100000:
        print('Property Id:',dict['id'],'Budgeted Amount:',dict['budgeted_amount'])
  
```
